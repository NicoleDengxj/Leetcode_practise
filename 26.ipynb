```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if not nums:  # Check if the list is empty
            return 0
        
        k = 1  # Start with the first unique element
        
        # Loop through the array starting from the second element
        for i in range(1, len(nums)):
            if nums[i] != nums[i - 1]:  # Compare with the previous element
                nums[k] = nums[i]  # Update the position for unique elements
                k += 1  # Move to the next position
        
        return k  # Return the count of unique elements

# Example usage:
nums1 = [1, 1, 2]
solution = Solution()
new_length1 = solution.removeDuplicates(nums1)
print(f"New length: {new_length1}, Modified array: {nums1[:new_length1]}")  # Output: 2, [1, 2]

nums2 = [0, 0, 1, 1, 1, 2, 2, 3, 3, 4]
new_length2 = solution.removeDuplicates(nums2)
print(f"New length: {new_length2}, Modified array: {nums2[:new_length2]}")  # Output: 5, [0, 1, 2, 3, 4]
```
