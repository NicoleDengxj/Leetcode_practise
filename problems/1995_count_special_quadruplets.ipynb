```python
from typing import List
class Solution:
    def countQuadruplets(self, nums: List[int]) -> int:
        n = len(nums)
        count = 0
        # Iterate over all quadruplet combinations (i, j, l, r)
        for i in range(n-3):
            for j in range(i+1, n-2):
                for l in range(j+1, n-1):
                    for r in range(l+1, n):
                        # Check if nums[i] + nums[j] + nums[l] equals nums[r]
                        if nums[i] + nums[j] + nums[l] == nums[r]:
                            count += 1
        return count
```

```python
nums = [1,2,3,4,2,3,6,8]
solution = Solution()
print(solution.countQuadruplets(nums))
```
