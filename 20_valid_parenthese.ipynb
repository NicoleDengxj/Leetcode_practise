```python
class Solution:
    def isValid(self,s:str)->bool:
        if s == "":
            return True
        if "()" in s:
            return self.isValid(s.replace("()",""))
        if "[]" in s:
            return self.isValid(s.replace("[]",""))
        if "{}" in s:
            return self.isValid(s.replace("[]",""))


```

```python
s = "[]()())"
solution = Solution()
print(solution.isValid(s))
```
