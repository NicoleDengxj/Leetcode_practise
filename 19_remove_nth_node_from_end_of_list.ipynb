```python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        if not head:
            return None
        
        # Create a dummy node to handle edge cases like removing the head node
        dummy = ListNode(0)
        dummy.next = head
        slow = dummy
        fast = dummy
        
        # Move fast pointer n+1 steps ahead, so there's exactly n nodes between slow and fast
        for _ in range(n + 1):
            fast = fast.next
        
        # Move both slow and fast until fast reaches the end of the list
        while fast:
            slow = slow.next
            fast = fast.next
        
        # Now, slow is pointing to the node before the one we want to remove
        slow.next = slow.next.next
        
        # Return the head of the modified list
        return dummy.next
```

```python
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> " if current.next else "\n")
        current = current.next
```

```python
def create_linked_list(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    return head
```

```python
# 示例链表: [1, 2, 3, 4, 5]
head = create_linked_list([1, 2, 3, 4, 5])

# 打印链表
print("原链表:")
print_linked_list(head)

# 移除倒数第2个节点
solution = Solution()
new_head = solution.removeNthFromEnd(head, 2)

# 打印新的链表
print("移除倒数第2个节点后的链表:")
print_linked_list(new_head)
```
