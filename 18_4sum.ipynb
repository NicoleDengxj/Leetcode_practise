```python
# my code in 06/10/2024
from typing import List
class Solution:
    def fourSum_2(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        n = len(nums)
        result = []
        i = 0


        while i<n-3:
            j = i+1
            while j<n-2:
                l = j+1
                r = n-1
                rem = target -nums[i] - nums[j]
                while l< r:
                    if nums[l] + nums[r] == rem:
                        result.append([nums[i],nums[j],nums[l],nums[r]])
                        while l+1<r and nums[l+1]==nums[l]:
                            l += 1
                        while r-1>l and nums[r-1] == nums[r]:
                            r -= 1
                        l +=1
                        r -=1
                    elif nums[l]+nums[r] < rem:
                        l +=1
                    elif nums[l]+nums[r] > rem:
                        r -=1
                while j<n-2 and nums[j+1] == nums[j]:
                    j +=1
                j +=1
            while i<n-3 and nums[i+1] == nums[i]:
                i +=1
            i +=1
        return result
    









    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        n = len(nums)
        result = []

        for i in range(n-3):
            for j in range(i+1,n-2):
                l = j + 1
                r = n - 1
                while(l<r):
                    sum = nums[i] +nums[j] +nums[l] +nums[r]
                    if sum == target:
                        result.append([nums[i], nums[j], nums[l], nums[r]])
                        l += 1
                        r -= 1
                        while l+1<r and nums[l] == nums[l+1]:
                            l += 1
                        while r-1 >l and nums[r-1] == nums[r]:
                            r -= 1
                    
                    elif sum > target:
                        r -= 1
                    elif sum < target:
                        l += 1
        final_result = []
        for item in result:
            if item not in final_result:
                final_result.append(item)
        return final_result
```
