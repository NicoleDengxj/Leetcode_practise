```python

class ListNode:
    def __init__(self,data):
        self.val = data  #链表的数据域
        self.next = None # 链表的指针域
```

```python
if __name__ == "__main__":
    head = ListNode(1)
    print(head.val)
    head.next = ListNode(123)
    pre = head.next
    print(pre.val)
    # head.next = ListNode(456)
    # pre = head.next
    head.next.next = ListNode(456)
    pre = head.next.next
    print(pre.val)
    
```
